```python
import pandas as pd 
import requests 
from bs4 import BeautifulSoup
```

```python
url ="https://www.flipkart.com/apple-watch-se-gps-2nd-gen-heart-rate-monitor-crash-detection-sleep-tracking/p/itm5484aad9a2dc0?pid=SMWGHWZ2RXWDWHED&fm=search-autosuggest&ppt=dynamic&ppn=dynamic&ssid=4d4asz1nts0000001701499519765"
```

```python
r =requests.get(url)
data =r.content
print(data)
```

```python
#prettify the codes of html page
soup =BeautifulSoup(data, "html.parser")
soup.prettify
```

```python
title =soup.title
print(type(soup))          #beautifulsoup
print(type(title))         #tag
print(type(title.string))  #navigablestring
```
